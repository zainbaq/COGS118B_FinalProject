```python
# convert class label to one-hot encoding
# convert class labels to on-hot encoding# conve 
Y = kf.utils.to_categorical(labels, num_classes=7,dtype='int32')
#Shuffle the dataset
x,y = shuffle(img_data,Y, random_state=2)
# Split the dataset
#X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=2)
```

```python
# start model building

model = kf.Sequential()
```

```python
# 1st add Convolution Layer
model.add(Conv2D(filters=96, input_shape=(227,227,3), kernel_size=(3,3),strides=(2,2), padding='same'))

#2nd add Convolution Layer
model.add(Conv2D(filters=96, kernel_size=(3,3),strides=(1,1), padding='same'))
model.add(Activation("relu"))

#3rd add Convolution Layer == 1st main 11 * 11 Conv

model.add(Conv2D(filters=96, kernel_size=(3,3),strides=(1,1), padding='same'))
model.add(Activation("relu"))

# local resp norm if not batch norm
model.add( BatchNormalization() )
# Pooling 
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2), padding='same'))

# Add Dropout
model.add(Dropout(0.5))
###########################################################################

# 2nd-main convo
model.add(Conv2D(filters=256, input_shape=(256,256,3),kernel_size=(5,5),strides=(2,2), padding='same'))
model.add(Activation("relu"))


# local resp norm if not batch norm
model.add( BatchNormalization() )

# Pooling 
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2), padding='same'))
# Add Dropout
model.add(Dropout(0.5))

################################################################
model.add(Conv2D(filters=384, kernel_size=(3,3),strides=(1,1), padding='same'))

model.add(Activation("relu"))

# local resp norm if not batch norm
model.add( BatchNormalization() )

#########################################################################

model.add(Conv2D(filters=384, kernel_size=(3,3),strides=(1,1), padding='same'))

model.add(Activation('relu'))

# local resp norm if not batch norm
model.add( BatchNormalization() )

#####################################################################################

model.add(Conv2D(filters=256, kernel_size=(3,3),strides=(1,1), padding='same'))

model.add(Activation('relu'))

# local resp norm if not batch norm
model.add( BatchNormalization() )

# Pooling 
model.add(MaxPooling2D(pool_size=(4,4), strides=(2,2), padding='same'))
# Add Dropout
model.add(Dropout(0.5))

# Passing it to a Fully Connected layer 1
model.add(Flatten())

# 1st Fully Connected Layer
model.add(Dense(512, input_shape=(227*227*3,)))
model.add(Activation("relu"))

# local resp norm if not batch norm
model.add( BatchNormalization() )
          
# Add Dropout to prevent overfitting
model.add(Dropout(0.5))
          
# Passing it to a Fully Connected layer 2

# 2nd Fully Connected Layer
model.add(Dense(512))
model.add(Activation("relu"))

# local resp norm if not batch norm
model.add( BatchNormalization() )
          
# Add Dropout to prevent overfitting
model.add(Dropout(0.5))     
          
# 3rd Fully Connected Layer
model.add(Dense(7))
model.add(Activation("relu"))
# Add Dropout
model.add(Dropout(0.5))

# Output Layer
model.add(Dense(7))
model.add(Activation("softmax"))


# (4) Compile 
model.compile(loss='categorical_crossentropy', optimizer=kf.optimizers.SGD(lr=0.01,momentum=0.9,decay=0.0005),\
 metrics=['accuracy'])

```

```python
model.summary()


# (5) Train
hist = model.fit(x, y, batch_size=128, epochs=10, verbose=1, \
validation_split=0.3, shuffle=True)
print(hist.history)
```
